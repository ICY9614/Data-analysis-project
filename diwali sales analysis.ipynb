```python
ax = sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender',y='Amount', data = sales_gen)
```

```python
#From above graph we can see that most of the buyers are females and even the purchasing power of females are greater than men 
```

```python
ax = sns.countplot(x='Age Group', hue = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data = sales_age)
```

```python
#From the above graph we can see that most of the buyers are of the age group between 26-35 yrs female
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Orders', data = sales_state)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Amount', data = sales_state)
```

```python
# From the above graph we can see that most of the orders & sales/amount are from U.P, Maharashtra and Karnataka
```

```python
ax = sns.countplot(x='Marital_Status', data = df, hue='Gender')

sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y='Amount', data = sales_state, hue='Gender')
```

```python
# From above the above graphs we can see that most of the buyers are married(woman) and they have high purchasing power
```

```python
ax = sns.countplot(x='Occupation', data = df)

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data = sales_occu)
```

```python
# From above the above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

```python
ax = sns.countplot(x='Product_Category', data = df)

sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x='Product_Category', y='Amount', data = sales_prod)
```

```python
# From above the above graphs we can see that most of the sold products are from Food , Clothing and Electronic category
```

```python
sales_prod = df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(x='Product_ID', y='Amount', data = sales_prod)
```

CONCLUSION:

Married women age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronic category.
